```python
import pandas as pd
```

```python
data = pd.read_csv("C:\\Users\\Utente\\Desktop\\Progetto Europee\\elezioni_europee_followers_leader_mv.csv", delimiter=";", encoding="unicode-escape")
```

```python
data
```

```python
salvini = data[data['Leader']=='Salvini'].reset_index()
salvini = salvini.drop("index", axis=1)
salvini = salvini[salvini['Ora']=='20:00'].reset_index()
salvini = salvini.drop("index", axis=1)
salvini
```

```python
for index, row in salvini.iterrows():
    if index>=1:
        salvini['Delta'].iloc[index]= ((salvini.iloc[index, 1]-salvini.iloc[index-1, 1])/salvini.iloc[index-1, 1])*100
    else: salvini['Delta']= 0
```
