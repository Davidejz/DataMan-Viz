```python
from __future__ import absolute_import, print_function
import tweepy
from tweepy import OAuthHandler
from tweepy import Stream
import json
```

```python
with open("C:\\Users\\Utente\\Desktop\\Data Science\\Data Management & Data Visualization\\Data Management\\secret.json", "r") as f:
    secret = json.load(f)
    consumer_key=secret["CONSUMER_KEY"]
    consumer_secret=secret["CONSUMER_SECRET"]

    access_token=secret["ACCESS_TOKEN"]
    access_token_secret=secret["ACCESS_TOKEN_SECRET"]

auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
search = api.user_timeline(screen_name = 'matteosalvinimi', count = 500, include_rts = True, tweet_mode= "extended") 
```

```python
with open("C:\\Users\\Utente\\Desktop\\Progetto Europee\\salvini_mod2.csv", "a", encoding="utf-8") as f:
    for s in search:
        f.write('{};{};{};{}\n'.format("Salvini", s.full_text, s.retweet_count, s.created_at))
```
